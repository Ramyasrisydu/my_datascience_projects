<b>Problem Statement</b><br>
This dataset includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family Mushroom drawn from The Audubon Society Field Guide to North American Mushrooms (1981). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like "leaflets three, let it be'' for Poisonous Oak and Ivy.

<b>Feature Engineering</b><br>
From the analysis above the following columns are not needed so we will drop them: gill-attachment, stalk-shape, stalk-surface-below-ring, stalk-color-below-ring, veil-type, veil-color, ring-number. so now we will create our feetures varable x excluding the mentioned columns:

```python
Del_cols = ['class','gill-attachment', 'stalk-shape', 'stalk-surface-below-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number'] 
X =data_set.copy()
for col in Del_cols:
    X = X.drop(col, axis = 1)
X=pd.get_dummies(X,columns=X.columns,drop_first=True)
X.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder

y = data_set['class']
Encoder_y=LabelEncoder()
y = Encoder_y.fit_transform(y)
y
```

<b>Score Metrics for Classification:</b><br>

Accuracy - Accuracy is the quintessential classification metric. It is pretty easy to understand. And easily suited for binary as well as a multiclass classification problem.<br>

Precision - Answers the following question: what proportion of predicted Positives is truly Positive?
Precision = (TP)/(TP+FP)<br>

Recall - Answers the following question: what proportion of actual Positives is correctly classified?
Recall = (TP)/(TP+FN)<br>

f1 score - A number between 0 and 1, the harmonic mean of precision and recall

```python
#train test split
from sklearn.model_selection import train_test_split

#metrics
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

#ML models
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.75, random_state = 101)
```

```python
models=[DecisionTreeClassifier(),RandomForestClassifier(),XGBClassifier(),LogisticRegression(),SVC()]
model_names=['DecisionTree','RandomForestClassifier','XGBoostClassifier','LogisticRegression','LinearSVM']

acc=[]
re=[]
pre=[]
f1=[]
d={}


for model in range(len(models)):
    clf=models[model]
    clf.fit(X_train,y_train)
    pred=clf.predict(X_test)
    acc.append(accuracy_score(pred,y_test))
    pre.append(precision_score(pred,y_test))
    re.append(recall_score(pred,y_test))
    f1.append(f1_score(pred,y_test))
    
d={'Modelling Algo':model_names,'Accuracy':acc, 'precision':pre, 'recall':re, 'f1 score':f1}
```

```python
metric_frame=pd.DataFrame(d)
metric_frame
```

```python
sns.barplot(y='Modelling Algo',x='Accuracy',data=metric_frame)
```
